# Import ground truth

```python
import json
```

```python
with open(path + 'regions/regions.json', 'r') as f:
    regions = json.load(f)
```

```python
from regional import many
truth = many([region['coordinates'] for region in regions])
```

```python
masks = truth.mask(stroke='deeppink', fill=None, base=mean.clip(0,4000) / 4000)
image(masks, size=10);
```
